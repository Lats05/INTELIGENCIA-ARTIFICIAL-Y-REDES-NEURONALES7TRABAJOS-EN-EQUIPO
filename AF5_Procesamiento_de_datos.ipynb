```python
# Primero importamos las librerías a utliizar durante el desarrollo de la actividad
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# También importamos drive para conservar el dataframe
from google.colab import drive
```

```python
# Cargamos los datos del archivo a trabajar
df= pd.read_csv ('/content/drive/MyDrive/Colab Notebooks/robot_sensors.csv')
print(df.head())
print("")
print("Dimensiones de la tabla:",df.shape)
```

```python
# Usamos las diferentes funciones para detectar valores atípicos
# Primero encontramos valores nulos y los corregimos
print("Valores Originales")
print(df.isnull().sum())
df_nonull=df.dropna()
print ("")
print ("Valores Corregidos")
print(df_nonull.isnull().sum())
print ("")
print ("Dimensiones de la tabla ANTES del filtrado (Filas,Columnas):", df.shape)
print("Dimensiones de la tabla DESPUÉS del filtrado (Filas,Columnas):", df_nonull.shape)
```

```python
# Ahora filtramos valores negativos y los mostramos
df_filtrada = df_nonull[df_nonull >= 0]
print (df_filtrada)
```

```python
# Después de filtrar, procedemos a normalizar los valores de cada columna para graficarlos después
# Normalización de la columna de distancia
dist= df_filtrada['Distancia (cm)']
dist_nor= (dist-np.min(dist))/(np.max(dist)-np.min(dist))
print("Valores de distancia normalizados")
print(dist_nor)
```

```python
# Normalización de la columna de velocidad
vel= df_filtrada['Velocidad (cm/s)']
vel_nor= (vel-np.min(vel))/(np.max(vel)-np.min(vel))
print("")
print("valores de velocidad normalizados")
print(vel_nor)
```

```python
# Normalización de la columna de temperatura
temp= df_filtrada['Temperatura (°C)']
temp_nor= (temp-np.min(temp))/(np.max(temp)-np.min(temp))
print("")
print("Valores de temperatura normalizados")
print(temp_nor)
```

```python
# Una vez normalizados los datos los graficaremos
tiempo= df_filtrada['Tiempo (s)']
# Gráfico de distancia
plt.scatter(tiempo, dist_nor, color='red', label="Distancia (cm)", s=6)
plt.title("Gráfico de Distancia")
plt.xlabel("Tiempo (s)")
plt.ylabel("Distancia (cm)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gráfico de velocidad
plt.scatter(tiempo, vel_nor, color='purple', label="Velocidad (cm/s)", s=6)
plt.title("Gráfico de Velocidad")
plt.xlabel("Tiempo (s)")
plt.ylabel("Velocidad(cm/s)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gráfico de temperatura
plt.scatter(tiempo, temp_nor, color='orange', label="Temperatura (°C)", s=6)
plt.title("Gráfico de Temperatura")
plt.xlabel("Tiempo (s)")
plt.ylabel("Temperatura (°C)")
plt.legend()
plt.grid(True)
plt.show()
```
